```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
X
```

```python
y
```

```python
plt.scatter(X,y)
```

```python
# Lets apply OLS
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X,y)
```

```python
print("b:",reg.coef_)
print("m:",reg.intercept_)
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color="green")
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and assume the starting value for intercept b = 0
y_pred = ((78.35063668 * X) + 0).reshape(4)
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color="green", label="OLS")
plt.plot(X, y_pred, color="#990000", label="b = 0")
plt.legend()
plt.show()
```

```python
m = 78.35063668
b = 0

# gradient = -2(y - mx - b)
loss_slope = -2 * np.sum(y - m * X.ravel() - b)
loss_slope
```

```python
# Lets take learning rate = 0.1
lr = 0.1

step_size = loss_slope * lr
step_size
```

```python
# Calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((78.35063668 * X) + b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color="green", label="OLS")
plt.plot(X, y_pred, color="#CC0000", label="b = {}".format(b))
plt.plot(X, y_pred1, color="#990000", label="b = 0")
plt.legend()
plt.show()
```

```python
b = 0
m = 78.35063668
lr = 0.01

epochs = 100

for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*X.ravel() - b)
    
    b = b - (lr * loss_slope)
    
    y_pred = m * X + b
    
    plt.plot(X,y_pred)

plt.scatter(X,y)
```
